```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# distplots

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

**Joint plots**

```python
sns.jointplot(x='total_bill',y='tip',data=tips) # combine two displots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

**Pairplot**

```python
# pairplot for all nummber colums
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
# rugplot() shows the number of dashes

sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill']) #kde kerneal density estimation plot
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical plots

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

**count plot**

```python
sns.countplot(x='sex',data=tips)
```

**Boxplot()**

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

**stripplot**

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='smoker')
```

**Swarmplot()**

```python
sns.swarmplot(x='day',y='total_bill',data=tips) # combination violinplot and stripplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

# Grid

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```
