```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x*2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Graph')
```

```python
# in subplot arguments are number of (rows,columns,plot-number)

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
#Object oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.3,0.3])
# left bottom width height

axes1.set_title('Bigger Title')
axes2.set_title('Smaller Title')
axes1.plot(x,y)
axes2.plot(y,x)
```

**Matplotlib part2**

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# axes became an array
for current_ax in axes:
  current_ax.plot(x,y)
  current_ax.set_title('Sub plots')
```

```python
axes[0].plot(x,y)
axes[0].plot(y,x)
```

Figure Size and DPI

```python
fig=plt.figure(figsize=(5,2),dpi=100) # dpi is the dots per inch or pixes
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,2)) # dpi is the dots per inch or pixes

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_png',dpi=100)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x**2,y,label='Xsquared')
ax.plot(x**3,y,label='Y cubed')
ax.legend(loc=10) # loc is 0 means best, 10 means center
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x**2,y,label='Xsquared')
ax.plot(x**3,y,label='Y cubed')
ax.legend(loc=(0.2,0.2))  # we can also mention place
# loc is 0 means best, 10 means center
```

```python
from matplotlib import colors
# Colors using matplotlib

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth='20') # linewidth is 20 here
```

```python
from matplotlib import colors
# Colors using matplotlib

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth='20',alpha=0.5)
```

```python
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('title')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
#Exercise 3

fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])

ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.set_xlabel('x')
ax2.set_ylabel('y')

ax1.plot(x,z)
ax2.plot(x,y)
```

```python
#Exercise 4

fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,lw=3,ls=':')
axes[1].plot(x,z)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(5,5))
axes[0].plot(x,y,lw=3,ls=':')
axes[1].plot(x,z)
plt.tight_layout()
```
